# Description

The Titanic dataset is a well-known dataset that provides information about the passengers who were on board the Titanic ship when it sank on April 15, 1912.The data includes demographic information such as the passengers' names, ages, genders, ticket class, fare, and cabin numbers, as well as information about their families and whether they survived or not.

# Problem Statement

The challenge is to use this information to build a predictive model that can accurately classify whether a passenger would have survived or not based on their characteristics without using libraries.

# Reading the dataset

```python
# Open the CSV file containing the Titanic dataset
with open('titanic.csv', 'r') as f:
    # Print the header row
    print(next(f))
    # Read the rest of the rows of data
    for row in f:
        # Split the row into columns
        columns = row.strip().split(',')
        # Print the columns
        print(columns)
```

Inference: 

This code opens the CSV file containing the Titanic dataset and reads through it line by line. It uses the next function to read and print the first line, which contains the column names. The next function returns the next item in the iterator, which in this case is the first line of the file.

After printing the header row, the code loops through the rest of the rows of data in the file. For each row, it splits the row into a list of columns using the split function, and then prints the columns. The strip function is used to remove any leading or trailing whitespace from each column.


# First 10 rows of the dataset

```python
# This line sets a variable n to 10, which represents the number of rows that the code will print.
n = 10  # The number of head rows to print

with open('titanic.csv', 'r') as f:
    dataset_content = f.readlines()

head_rows = dataset_content[:n]

for row in head_rows:
    columns = row.strip().split(',')
    print(columns)
```

The next line head_rows = dataset_content[:n] selects the first n rows of the dataset and stores them in a new list called head_rows.

The for loop then iterates over each row in head_rows, splits the row into a list of columns using the split() method, and prints the columns.

```python
# This line sets a variable n to 10, which represents the number of rows that the code will print.
n = 10  # The number of tail rows to print

with open('titanic.csv', 'r') as f:
    dataset_content = f.readlines()

tail_rows = dataset_content[-n:]

for row in tail_rows:
    columns = row.strip().split(',')
    print(columns)
```

# Checking for missing values along with row information

```python
import csv

# Open the Titanic dataset CSV file
with open('titanic.csv', 'r') as file:

    # Create a CSV reader object
    reader = csv.reader(file)

    # Extract the header row
    header = next(reader)

    # Loop through the data rows and check for missing values
    for row in reader:
        for i in range(len(header)):
            if row[i] == '':
                print(f"Missing value found in column '{header[i]}' at row {reader.line_num}")
```

# Checking for missing values 

```python
import csv

# Open the Titanic dataset CSV file
with open('titanic.csv', 'r') as file:

    # Create a CSV reader object
    reader = csv.reader(file)

    # Extract the header row
    header = next(reader)

    # Initialize a dictionary to store the number of missing values per column
    missing_values = {column: 0 for column in header}

    # Loop through the data rows and count missing values in each column
    for row in reader:
        for i, value in enumerate(row):
            if value == '':
                missing_values[header[i]] += 1

    # Print the columns with missing values and their respective counts
    print("Columns with missing values:")
    for column, count in missing_values.items():
        if count > 0:
            print(f"{column}: {count}")
```
